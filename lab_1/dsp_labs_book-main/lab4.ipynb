# Мел-фильтры
Одними из наиболее популярных акустических признаков являются Filter Banks (fbanks). 
fbanks вычисляются применением нескольких (количество фильтров = количество fbanks) треугольных фильтров к мел-спектрограмме. Чтобы не делать два действия со спектрограммой, переход к мел-шкале и применение фильтров в мел-шкале можно заменить на перевод мел-фильтров в Герц-шкалу и применение их к Герц-спектрограмме.

### Мел-фильтры

Реализуйте функцию вычисления fbank.
Помимо лекций описание формул можно посмотреть тут: https://studbooks.net/2037683/informatika/raschyot_filtrov. 

```python
def mel_filters_pure(sr, n_fft, n_mels):
    # функция построения треугольных мел-фильтров в герц-шкале
    # sr - sample rate
    # n_fft - length of the FFT window 
    # n_mels - number of filters
    # return mel filters matrix. [n_mel, n_fft]

    min_mel_freq = 0
    max_mel_freq = 2595 * np.log10(1 + (sr / 2) / 700)
    mel_centers = np.linspace(min_mel_freq, max_mel_freq, n_mels + 2)
    mel_centers_hz = 700 * (np.exp(mel_centers / 1127) - 1)
    mel_filters = np.zeros((n_mels, n_fft // 2 + 1))
    bin_indices = mel_centers_hz / (sr / n_fft)
    for i in range(1, n_mels + 1):
        left_edge = int(bin_indices[i - 1])
        center = int(bin_indices[i])
        right_edge = int(bin_indices[i + 1])
        for j in range(left_edge, center):
            mel_filters[i - 1, j + 1] = (j + 1 - bin_indices[i - 1]) / (bin_indices[i] - bin_indices[i - 1])
        for j in range(center, right_edge):
            mel_filters[i - 1, j + 1] = (bin_indices[i + 1] - (j + 1)) / (bin_indices[i + 1] - bin_indices[i])
    return mel_filters
```

Обратите внимание что реализации базовых функций таких как вычисления fbank могут отличаться в различных широко используемых библиотеках. Как правило это не критично для большинства практических задач, но бывают сложности при поиске ошибок и сравнении реализаций. 
Реализуйте функцию вычисления fbank по формулам библиотеки librosa по аналогии с открытым исходным кодом. Формулы, используемые в библиотеке librosa, отличаются от используемых в лекциях и некоторых других источниках. Обратите внимание на использование np.diff и np.substract, а также на нормализацию slaney. https://librosa.org/doc/latest/_modules/librosa/filters.html#mel

```python
def mel_filters_librosa(sr, n_fft, n_mels):
    # функция построения треугольных мел-фильтров в герц-шкале
    # sr - sample rate
    # n_fft - length of the FFT window 
    # n_mels - number of filters
    # return mel filters matrix. [n_mel, n_fft]

    n_mels = int(n_mels)
    weights = np.zeros((n_mels, int(1 + n_fft // 2)), dtype=np.float32)
    fmax = float(sr) / 2
    mel_f = librosa.core.convert.mel_frequencies(n_mels + 2, fmin=0.0, fmax=fmax, htk=True)
    fftfreqs = librosa.core.convert.fft_frequencies(sr=sr, n_fft=n_fft)

    fdiff = np.diff(mel_f)
    ramps = np.subtract.outer(mel_f, fftfreqs)

    for i in range(n_mels):
        lower = -ramps[i] / fdiff[i]
        upper = ramps[i + 2] / fdiff[i + 1]
        weights[i] = np.maximum(0, np.minimum(lower, upper))

    enorm = 2.0 / (mel_f[2 : n_mels + 2] - mel_f[:n_mels])
    weights *= enorm[:, np.newaxis]
    return weights
```

```python
assert mel_filters_librosa(32, 46, 4).shape == (4, 24) and \
    mel_filters_librosa(65, 45, 5).shape == (5, 23), "Wrong shape"
assert np.allclose(mel_filters_librosa(16, 8, 4), librosa.filters.mel(sr=16, n_fft=8, n_mels=4, htk=True))
assert np.allclose(mel_filters_librosa(8600, 512, 40), librosa.filters.mel(sr=8600, n_fft=512, n_mels=40, htk=True))
print("All ok!")
```

Можно заметить, что есть небольшое расхождение в результатах формул. Для большенства практических задач это не существенно, но эти различия полезно знать при собственных экспериментах с аудио фичами а так же в случаях отладки и поиске ошибок.

Далее будем использовать функцию *mel_filters_librosa*

```python
def get_fbanks(wav: np.ndarray, sr: int, window_ms=25, step_ms=10, n_fbanks=40):
    # wav - input signal
    # sr - sample rate
    # window_ms - window length in milliseconds
    # step_ms - stft step in milliseconds
    # n_fbanks - number of filters
    # return fbank matrix [n_fbanks, time]
    
    n_fft = window_ms * sr//1000
    hop_length = step_ms * sr//1000
    sft = librosa.stft(wav, n_fft=n_fft, hop_length=hop_length)
    magnitude = np.abs(sft) ** 2
    mels = mel_filters_librosa(sr, n_fft, n_fbanks)
    return mels.dot(magnitude)
```

```python
def test_fbank(wav, sr, window_ms=25, step_ms=10, n_fbanks=40):
    n_fft = window_ms * sr//1000
    hop_length = step_ms * sr//1000
    fbanks_lib = librosa.feature.melspectrogram(y=wav, sr=sr, n_fft=n_fft, hop_length=hop_length, n_mels=n_fbanks, htk=True)
    fbanks = get_fbanks(wav, sr, window_ms=window_ms, step_ms=step_ms, n_fbanks=n_fbanks)

    if fbanks_lib.shape != fbanks.shape:
        print("TEST FAILED")
        print(f"Shape {fbanks_lib.shape} != {fbanks.shape}")
        return 1

    if not np.allclose(fbanks_lib, fbanks):
        print('TEST PASSED BUT WITH CALCULATION ERROR')
        print(f"Average diff is {np.mean(np.abs(fbanks_lib - fbanks))}")
        return np.mean(np.abs(fbanks_lib - fbanks))
        
    print("TEST PASSED")
    return 0

assert test_fbank(wav[:sr*1], sr) <= 0.0003, "1 sec wav test failed"
assert test_fbank(wav, sr) <= 0.0003, "All wav test failed"
print("All ok!")
```

Реализовать вычисление [mfcc](https://en.wikipedia.org/wiki/Mel-frequency_cepstrum)



```python
def get_mfcc(wav: np.ndarray, sr: int, window_ms=25, step_ms=10, n_mfcc=13):
      # wav - input signal
    # sr - sample rate
    # window_ms - window length in milliseconds
    # step_ms - stft step in milliseconds
    # n_mfcc - number of filters
    # return mfcc matrix [n_mfcc, time]
    
    fbanks = get_fbanks(wav, sr, window_ms, step_ms, 128)
    fbanks = librosa.power_to_db(fbanks)
    mfcc = np.zeros((n_mfcc, fbanks.shape[1]))
    for i in range(fbanks.shape[1]):
        mfcc[:,i] = scipy.fft.dct(fbanks[:,i], norm='ortho')[:n_mfcc]
    return mfcc
```

```python
def load_yn_dataset(directory):
    X, labels = [], []
    bad_files = set(["0_1_0_1_0_0_0_0"])
    for f in glob(directory +"/*.wav"):
        name = os.path.basename(f)[:-4]
        if name in bad_files:
            continue
        y = [int(l) for l in name.split("_")]
        x, _ = librosa.load(f)
        X.append(x)
        labels.append(y)
        
    return X, labels
        
X, Y = load_yn_dataset("waves_yesno/")
```

```python
# YOUR CODE HERE

train_VA = []  # List[np.ndarray]. 1 - Voice, 0 - silence
test_VA = []  # List[np.ndarray]. 1 - Voice, 0 - silence

# Параметры обработки аудио
window_length_ms = 120  # Определено экспериментально как наиболее эффективная длина окна
step_length_ms = 50  # Определено экспериментально как наиболее эффективная длина шага
threshold = 195  # Определено экспериментально как наиболее эффективный порог для определения наличия голоса в кадре

# Расчет размеров шагов в выборках
hop_step_samples = int(sr / 1000 * step_length_ms)
window_step_samples = int(sr / 1000 * window_length_ms)

# Обработка тренировочного набора данных
for audio in X_train:
    audio_mfcc = get_mfcc(audio, sr, window_length_ms, step_length_ms)
    energy = audio_mfcc[1]
    voice_activity = np.zeros(energy.shape)
    voice_activity[energy > threshold] = 1
    
    # Расширение меток голосовой активности до исходной частоты дискретизации
    real_voice_activity = np.zeros(hop_step_samples * (voice_activity.shape[0] - 1) + window_step_samples)
    for i in range(voice_activity.shape[0]):
        real_voice_activity[i * hop_step_samples:i * hop_step_samples + window_step_samples] = voice_activity[i]
    
    # Обрезка до длины исходного аудиосигнала
    real_voice_activity = real_voice_activity[:len(audio)]
    train_VA.append(real_voice_activity)

# Обработка тестового набора данных
for audio in X_test:
    audio_mfcc = get_mfcc(audio, sr, window_length_ms, step_length_ms)
    energy = audio_mfcc[1]
    voice_activity = np.zeros(energy.shape)
    voice_activity[energy > threshold] = 1
    
    # Расширение меток голосовой активности до исходной частоты дискретизации
    real_voice_activity = np.zeros(hop_step_samples * (voice_activity.shape[0] - 1) + window_step_samples)
    for i in range(voice_activity.shape[0]):
        real_voice_activity[i * hop_step_samples:i * hop_step_samples + window_step_samples] = voice_activity[i]
    
    # Обрезка до длины исходного аудиосигнала
    real_voice_activity = real_voice_activity[:len(audio)]
    test_VA.append(real_voice_activity)

print(len(train_VA))
print(len(test_VA))
```

```python
def test_VAD(VOICE, VA, Y, min_acc=0.95):
    def check_ali(ali, num_words):
        diff = ali[1:]  - ali[:-1]
        if diff.sum()!=0:
            print("VAD detected speech at the beginning (or end) of audio")
            return -1 
        if not (diff > 0).sum() == (diff > 0).sum() == num_words:
            return -2
        return 0
    
    bad_i = []
    err =0
    for i, (ali, y) in enumerate(zip(VA, Y)):
        assert len(VOICE[i]) == len(ali), "shapes doesn't match"
        if check_ali(ali, len(y)) != 0:
            bad_i.append(i)
            err+=1
    acc = (len(Y) - err)/len(Y)
    print(f"Accuracy {acc}")
    if len(bad_i) >0:
        print(f"ids with error {bad_i}")
    assert acc >= min_acc , f"Accuracy must be >= {min_acc}"
    print("Test passed")

test_VAD(X_train, train_VA, Y_train)
test_VAD(X_test, test_VA, Y_test)
print("VAD ok!")
```

# Задание *6 (2 балла)
Обучите классификатор, определяющий, какое слово было сказано. Используйте VAD для разбиения входных файлов на отдельные слова. Классификацию можно сделать, например, с помощью SVM по усредненным признаки выделеных VAD'ом слов. Или любым другим удобным для вас способом. Добейтесь точности больше 0.95

```python
# YOUR CODE HERE

def extract_word_fragments(audio_list, vad_list):
    """Извлекает фрагменты слов из аудио"""
    word_fragments = []
    total_length = 0
    for i, audio in enumerate(audio_list):
        start_index, end_index = 0, 0
        p = 0
        while p < len(audio):
            last_value = vad_list[i][p - 1] if p - 1 >= 0 else 0
            next_value = vad_list[i][p + 1] if p + 1 < len(audio) else 0
            if last_value == 0 and vad_list[i][p] != 0:
                start_index = p
            if vad_list[i][p] != 0 and next_value == 0:
                end_index = p + 1
                word_fragments.append(audio[start_index:end_index])
                total_length += end_index - start_index
            p += 1
    average_length = int(total_length / len(word_fragments))
    return word_fragments, average_length

word_fragments_train, average_length = extract_word_fragments(X_train, train_VA)

def extract_mfcc_features(word_list, average_length, window_ms, step_ms):
    """Извлекает признаки MFCC для каждого слова"""
    mfcc_features = []
    for i, word in enumerate(word_list):
        if len(word) > average_length:
            word = word[:average_length]
        if len(word) < average_length:
            word = np.append(word, np.zeros(average_length - len(word)))
        word_mfcc = get_mfcc(word, sr, window_ms, step_ms)
        mfcc_features.append(word_mfcc.flatten())
    mfcc_features = np.array(mfcc_features)
    return mfcc_features

y_train_labels = np.array(Y_train).flatten()
# Каждый набор признаков MFCC имеет размерность 12x13, поэтому уплощаем его до 156 признаков
mfcc_features_train = extract_mfcc_features(word_fragments_train, average_length, 50, 25)

from sklearn import svm
svm_classifier = svm.SVC()
# mfcc_features_train имеет размерность 376x156, y_train_labels имеет размерность 376
svm_classifier.fit(mfcc_features_train, y_train_labels)
```
